1. Machine Learning adalah cabang dari kecerdasan buatan yang berfokus pada pengembangan sistem yang dapat belajar dari data, mengidentifikasi pola, dan membuat keputusan dengan sedikit atau tanpa intervensi manusia. Tujuan utama dari machine learning adalah untuk mengembangkan algoritma yang dapat mengidentifikasi pola yang tersembunyi dalam data dan menggunakan pola tersebut untuk membuat prediksi atau mengambil tindakan tertentu. Ada beberapa pendekatan dalam machine learning, termasuk:

    a) Supervised Learning: Model diberikan contoh data yang sudah dilabeli (dipasangkan dengan output yang diharapkan) dan harus mempelajari hubungan antara input dan output. Tujuannya adalah untuk membuat prediksi yang akurat tentang data baru.

    b) Unsupervised Learning: Model diberikan data yang tidak memiliki label dan diminta untuk menemukan struktur atau pola dalam data tersebut sendiri. Tujuan utamanya adalah untuk menemukan pola tersembunyi atau struktur yang mungkin tidak terlihat oleh manusia.

    c) Reinforcement Learning: Model belajar melalui interaksi dengan lingkungan dan menerima umpan balik berdasarkan tindakan yang diambilnya. Tujuannya adalah untuk memaksimalkan reward yang diterima oleh model dari lingkungan.

    Dalam proses pembelajaran, model machine learning biasanya melewati beberapa tahap, termasuk preprocessing data, pemilihan model, training, evaluasi, dan fine-tuning untuk meningkatkan kinerja. Machine learning telah menemukan aplikasi luas dalam berbagai domain, termasuk pengenalan wajah, pengenalan suara, deteksi pola dalam data keuangan, diagnosa medis, dan banyak lagi.

3.	Dalam penerapan Machine Learning, terdapat beberapa macam taxonomi atau klasifikasi yang dapat digunakan untuk mengelompokkan metode dan pendekatan. Berikut adalah beberapa taxonomi umum dalam Machine Learning:

    a) Berdasarkan Tujuan Pembelajaran:
    - Supervised Learning: Model belajar dari data yang sudah dilabeli dan membuat prediksi atau keputusan berdasarkan contoh yang sudah diberi label.
    - Unsupervised Learning: Model belajar dari data yang tidak memiliki label dan mencoba untuk menemukan pola atau struktur yang tersembunyi dalam data.
    - Semi-Supervised Learning: Model belajar dari campuran data yang dilabeli dan tidak dilabeli, yang memanfaatkan informasi yang tersedia dalam data yang tidak dilabeli untuk meningkatkan kinerja model.
    - Reinforcement Learning: Model belajar melalui interaksi dengan lingkungan, menerima umpan balik dalam bentuk reward atau hukuman, dan memperbarui strategi atau keputusan berdasarkan umpan balik tersebut.
    
    b) Berdasarkan Jenis Output:
    - Regression: Model memprediksi nilai kontinu, seperti harga saham atau suhu.
    - Classification: Model memprediksi kelas atau label dari data, seperti klasifikasi gambar menjadi kucing atau anjing.
    - Clustering: Model mengelompokkan data ke dalam kelompok atau klaster berdasarkan kesamaan mereka.
    - Dimensionality Reduction: Model mengurangi dimensi data dengan mempertahankan informasi penting dan menghilangkan noise atau redundansi.
    
    c) Berdasarkan Pendekatan Model:
    - Parametric Learning: Model memiliki parameter yang tetap dan mempelajari fungsi yang paling cocok dengan parameter tersebut, seperti regresi linier.
    - Non-parametric Learning: Model tidak memiliki parameter tetap dan fleksibel dalam menyesuaikan dengan data, seperti K-Nearest Neighbors.
    - Instance-based Learning: Model mempelajari pola langsung dari contoh-contoh dalam data tanpa proses pelatihan yang eksplisit.
    - Model-based Learning: Model mempelajari struktur data dan membuat generalisasi berdasarkan model probabilistik atau statistik.
    
    d) Berdasarkan Metode Pembelajaran:
    - Batch Learning: Model belajar dari seluruh dataset sekaligus.
    - Online Learning: Model belajar secara bertahap dari data yang masuk satu per satu atau dalam batch kecil.
    - Transfer Learning: Model memanfaatkan pengetahuan yang diperoleh dari tugas atau domain yang berbeda untuk meningkatkan kinerja pada tugas tertentu.
    - Meta Learning: Model belajar cara belajar sendiri, seperti mengoptimalkan arsitektur atau hiperparameter model.

    Taxonomi ini membantu para peneliti dan praktisi Machine Learning untuk memahami berbagai jenis pendekatan, teknik, dan aplikasi, serta memilih pendekatan yang paling sesuai untuk masalah atau tugas yang mereka hadapi.

````python
import pandas as pd

data = {
    'Hari': ['Senin', 'Selasa', 'Rabu', 'Kamis', 'Jumat', 'Sabtu', 'Minggu'],
    'Datang': [2, 3, 4, 1, 1, 5, 2],
    'Biaya': [30000*2, 35000*3, 25000*4, 15000*1, 20000*1, 30000*5, 35000*2],
    'Mahasiswa': ['Ani', 'Budi', 'Jono', 'Lono', 'Joni', 'Ani', 'Budi']
}

dataset = pd.DataFrame(data)
rata_rata_datang = dataset['Datang'].mean()
biaya_tertinggi_hari = dataset.loc[dataset['Biaya'].idxmax(), 'Hari']
biaya_tertinggi = dataset['Biaya'].max()
hari_biaya_lebih_110000 = dataset[dataset['Biaya'] > 110000]['Hari'].tolist()
mahasiswa_terbanyak = dataset['Mahasiswa'].value_counts().idxmax()
datang_minggu = dataset.loc[dataset['Hari'] == 'Minggu', 'Mahasiswa'].values[0]
biaya_terendah = dataset['Biaya'].min()
frekuensi_tertinggi = dataset['Datang'].max()
frekuensi_terendah = dataset['Datang'].min()

print(f"a) {rata_rata_datang:.2f}")
print(f"b) {biaya_tertinggi_hari}")
print(f"c) {', '.join(hari_biaya_lebih_110000)}")
print(f"d) {mahasiswa_terbanyak}")
print(f"e) {datang_minggu}")
print(f"f) Tertinggi: {biaya_tertinggi}, Terendah: {biaya_terendah}")
print(f"g) {frekuensi_terendah}")
````
